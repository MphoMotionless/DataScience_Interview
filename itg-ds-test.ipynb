```python
# since sklearn 0.22, you can use sklearn.metrics.plot_confusion_matrix()
def plot_confusion_matrix(cm, classes, title='confusion_matrix', cmap=plt.cm.gray,):
    
    plt.imshow(cm,interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    plt.tight_layout()
    plt.ylabel('True_label')
    plt.xlabel('Predicted label')
#     fig = plt.figure(figsize=(15,15))
    
    
```

```python
cm_plot_labels = ['Downstairs', 'Jogging', 'Sitting', 'Standing', 'Upstairs', 'Walking']
plot_confusion_matrix(cm, classes=cm_plot_labels, title='Confusion Matrix')
```
